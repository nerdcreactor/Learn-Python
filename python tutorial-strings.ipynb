Datatypes in python 

Basically, a datatype is a metadata of data we entered in our program. A datatype is a description of what kind of values data contains and what type of operations can be perfomed on this data.
In python everything is a object so datatypes are actually classes and variables are instances of these classses.

```python
var_1 = 123
var_2 = 'Ayush is very amazing'
var_3 = ['text', 123]
var_4 = {'name':'john', 'code': 2356 }
var_5 = (1, 2, 3, 4, 5)
var_6 = {'this is not a dictionary', 'its something different', 'notice the difference'}

var_7 = True
```

```python
#we can use type() method to check the data type

print(type(var_1))
print(type(var_2))
print(type(var_3))
print(type(var_4))
print(type(var_5))
print(type(var_6))
print(type(var_7))
```

```python
#len() function counts the characters in a string

current_status = 'You are doing great! Keep learning'
char_in_str = len(current_status)
print(char_in_str)
```

```python
#or you can also print it as

print('Total number of characters in the string are: {}'.format(char_in_str))
```

```python
#index in string2

my_name = 'Ayush'
print(my_name[0])
print(my_name[1])
print(my_name[2])
print(my_name[3])
print(my_name[4])
```

```python
#printing a range of characters in strings and slicing

print(my_name[0:1])
print(my_name[0:2])
print(my_name[1:3])
print(my_name[2:4])
print(my_name[0:5])
print(my_name[:])

      
```
